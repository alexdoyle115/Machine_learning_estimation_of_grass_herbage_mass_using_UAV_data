## 4.0 Linear Model
---
Some of the analysis was completed in R primarily exploring the linear relationship of the dependent variable (compressed sward height) and the independent variables (the model estimations and the results of the vegitation index transformations).

```python
%reload_ext rpy2.ipython
```

```python
%%R

# r imports 
library(openxlsx)
library(visreg)

r_data <- read.xlsx('predictions.xlsx')
```

```python
%%R

pairs(r_data[,c('small_cnn', 'green', 'vndvi')])
round(cor(r_data[,c('small_cnn', 'green', 'vndvi')]), 2)
```

```python
%%R

# the linear model
lm0 <- lm(formula = label ~ small_cnn + log(vndvi) + green, data = r_data)

anova(lm0)
summary(lm0)
```

```python
%%R

# plotting the linear regression
visreg(lm0, ylab = "Compressed Sward Height (mm)")
```
